# Seleciona Produtos

```python
import numpy as np
import pandas as pd
import mysql.connector
from fbprophet import Prophet

#Conecta com banco MySQL
try:
    connection = mysql.connector.connect(host='localhost',
                                         database='projeto',
                                         user='root',
                                         password='12345678')
    
    #produtos com código e tipo de publicação 1 (impresso)
    sql_select_Query = " select sk_produto \
                        from dim_produto as pd \
                        where pd.sk_produto > 0 \
                        and cd_mae = '00117038' "
    
    cursor = connection.cursor()
    cursor.execute(sql_select_Query)
    
    #carrega dados
    records = cursor.fetchall()

    data_sql = pd.DataFrame(records)

except mysql.connector.Error as e:
    print("Error reading data from MySQL table", e)
finally:
    if connection.is_connected():
        connection.close()
        cursor.close()
        print("MySQL connection is closed")
        
df_prod = data_sql.rename(columns={0: "sku"})
        
df_prod.info()
```

```python
#Numero de períodos para predição
n_periods = 24

for ind in df_prod.index:
    
    nProduto = df_prod['sku'][ind]
    
    print("produto " + str(int(nProduto)) )
    
    #Conecta com banco MySQL
    try:
        connection = mysql.connector.connect(host='localhost',
                                             database='projeto',
                                             user='root',
                                             password='12345678')


        sql_select_Query = " select ds_data_sem_hora \
                            ,isbn_produto \
                            ,val_venda \
                            from fato_venda as ft \
                            , dim_tempo as tp \
                            , dim_produto as pd \
                            where tp.sk_tempo = ft.sk_tempo \
                            and pd.sk_produto = ft.sk_protudo \
                            and pd.sk_produto = " + str(int(nProduto))           

        cursor = connection.cursor()
        cursor.execute(sql_select_Query)

        #carrega dados
        records = cursor.fetchall()

        data_sql = pd.DataFrame(records)

    except mysql.connector.Error as e:
        print("Error reading data from MySQL table", e)
    finally:
        if connection.is_connected():
            connection.close()
            cursor.close() 
            print("MySQL connection is closed")

    if len(data_sql)>0:
       
        print("data_sql>0")

        #ajusta dados para a predição
        df_mov = data_sql.copy()

        #adiciona nome nas colunas
        df_mov = df_mov.rename(columns={0: "date",1:"isbn", 2: "value"})

        #transforma para datetime
        df_mov["date"] = pd.to_datetime(df_mov["date"])

        #ordena
        df_mov = df_mov.sort_values('date')

        #soma quantidade em datas iguais
        df_mov = df_mov.groupby('date')['value'].sum().reset_index()

        #transforma a data em indice
        df_mov = df_mov.set_index('date')

        #agrupa na soma do mês
        df_mov = df_mov['value'].resample('MS').sum()

        if len(df_mov)>= 24:
            
            print("df_mov >= 24")

            df_mov = pd.to_numeric(df_mov, downcast='signed')
            df_mov = df_mov.where(df_mov<df_mov.quantile(0.95),df_mov.quantile(0.95))

            predicao = []

            #Quantida de períodos futuros
            fut_per = int(n_periods/2)

            df_prophet = df_mov.copy()

            #Ajust dados
            df_prophet = df_prophet.to_frame()

            #Reinicia indice
            df_prophet = df_prophet.reset_index()

            #adiciona nome nas colunas
            df_prophet = df_prophet.rename(columns={"date":"ds","value":"y"})

            #Fit modelo Prophet
            fitProphet = Prophet()

            fitProphet.fit(df_prophet)

            #Forecast Prophet
            future = fitProphet.make_future_dataframe(periods=fut_per, freq='MS')

            predProphetfull = fitProphet.predict(future)

            predicao = predProphetfull[['ds','yhat']]

            predicao = predicao.set_index('ds')

            predicao = predicao.rename(columns={"yhat": "value"})

            predicao = predicao.iloc[-24:]

            predicao['no_modelo'] = 'Prophet'

            predicao = predicao.round(2)
            
            df_export = predicao.copy()
            
            df_export["sk_tempo"] = df_export.index.strftime('%Y%m%d').astype(int)
            df_export["sk_produto"] = int(nProduto)

            df_export = df_export.replace(np.nan,0)
            
            df_export.reset_index(drop=True)

            try:

                print("Atualiza fato simulacao")

                # Connect to the database
                connection = mysql.connector.connect(host='localhost',
                                                     database='projeto',
                                                     user='root',
                                                     password='12345678')

                cursor = connection.cursor()

                # Create a new record
                query = " DELETE FROM projeto.fato_simulacao WHERE sk_produto = " + str(int(df_prod['sku'][ind])).strip()

                cursor.execute(query)

                # connection is not autocommit by default. So we must commit to save our changes.
                connection.commit()

                # Create a new record
                #query = """INSERT INTO projeto.fato_predicao (sk_tempo, sk_protudo, qtd_predicao, qtd_acumula) VALUES(%s, %s, %s, %s) ON DUPLICATE KEY UPDATE qtd_predicao=%s, qtd_acumula=%s"""
                #cursor.execute(query, (df_export.sk_tempo[i], df_export.sk_protudo[i], df_export.PROPHET[i], df_export.Ac_PROPHET[i], df_export.PROPHET[i], df_export.Ac_PROPHET[i]))

                query = """INSERT INTO projeto.fato_simulacao (sk_tempo, sk_produto, no_modelo, qtd_predicao, qtd_acumula) VALUES(%s, %s, %s, %s, %s) ON DUPLICATE KEY UPDATE qtd_predicao=%s, qtd_acumula=%s"""

                for row in df_export.index:
                                        
                    nTempo = df_export['sk_tempo'][row].item()
                    nProduto = df_export['sk_produto'][row].item()
                    cModelo = df_export['no_modelo'][row]
                    nQtdInsert = df_export['value'][row].item()
                    nAcmInsert = 0
            
                    #Grava dados da simulacao
                    cursor.execute(query, (nTempo, nProduto, cModelo, nQtdInsert, nAcmInsert, nQtdInsert, nAcmInsert))
                    
                # connection is not autocommit by default. So we must commit to save our changes.
                connection.commit()
                
                # Execute query
                #sql = "SELECT * FROM projeto.fato_simulacao WHERE sk_produto = " + str(int(nProduto)).strip()

                #cursor.execute(sql)

                # Fetch all the records
                #result = cursor.fetchall()
                
                #print(result)

            finally:

                # close the database connection using close() method.
                connection.close()

print("end of cron")
```
